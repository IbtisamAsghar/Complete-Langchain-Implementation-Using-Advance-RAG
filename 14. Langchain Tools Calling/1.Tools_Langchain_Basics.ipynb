# Building the custom tool

```python
from langchain_core.tools import tool
```

```python
# step 1 - create a function
def finding_factorial (a) :
  """Finds the factorial of the number"""
  result = 1
  for i in range(1,a+1) :
    result *= i
  return result
```

```python
# step 2 - add type hints

def finding_factorial(a : int) -> int :
  """Finds the factorial of the number"""
  result = 1
  for i in range(1,a+1) :
    result *= i
  return result
```

```python
# step 3 - add tool decorator
@tool
def finding_factorial(a : int) -> int :
  """Finds the factorial of the number"""
  result = 1
  for i in range(1,a+1) :
    result *= i
  return result
```

```python
result = finding_factorial.invoke({"a":5})
```

```python
print(result)
```

```python
print(finding_factorial.name)
print(finding_factorial.description)
print(finding_factorial.args)
```

```python
print(finding_factorial.args_schema.model_json_schema())
```

# Method 2 - Using Structured Tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel , Field
```

```python
from typing_extensions import Required

class factorial_number (BaseModel) :
  a : int = Field(Required = True , description= "Enter the number to calculate he factorial")
```

```python
def finding_factorial(a : int) -> int :
  """Finds the factorial of the number"""
  result = 1
  for i in range(1,a+1) :
    result *= i
  return result
```

```python
fact_tool = StructuredTool.from_function(
    func = finding_factorial,
    name = "Finding factorial of number ",
    description = "Finding the factorial of the given number ",
    args_schema=factorial_number
)
```

```python
result = fact_tool.invoke({'a':5})
```

```python
print(result)
```

```python
print(fact_tool.name)
print(fact_tool.description)
print(fact_tool.args)
```

```python
print(fact_tool.args_schema.model_json_schema())
```

# Method 3 - Using base Tool Class

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
from typing_extensions import Required
class factorial_number (BaseModel) :
  a : int = Field(Required = True , description= "Enter the number to calculate he factorial")
```

```python
class MultiplyTool(BaseTool) :
  name : str= "Factorial Finder"
  description : str = "Find the factorial of the number"
  args_schema : type[BaseModel] = factorial_number

  def _run(self,a : int) -> int :
    result = 1
    for i in range(1,a+1) :
      result *= i
    return result
```

```python
factorial_tool = MultiplyTool()
```

```python
result = factorial_tool.invoke({"a":5})
```

```python
print(result)
```

```python
print(fact_tool.name)
print(fact_tool.description)
print(fact_tool.args)
```
