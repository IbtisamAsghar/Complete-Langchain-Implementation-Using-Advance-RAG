## Lord of Retrievers (LOTR)

```python
from langchain.retrievers.merger_retriever import MergerRetriever
```

```python
lotr = MergerRetriever(retrievers=[retriever_harrypotter , retriever_got])
```

# Filter the ouput

```python
from langchain.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter, # same retrieved document will be removed
)
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.document_transformers import LongContextReorder
```

```python
filter = EmbeddingsRedundantFilter(embeddings=hf_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
```

```python
compression_retriever = ContextualCompressionRetriever(
    base_compressor = pipeline,
    base_retriever =lotr,
    search_kwargs = {"k":3 , "include_metadata":True}
)
```

```python
llm_model = ChatOpenAI(model_name="gpt-3.5-turbo")
```

```python
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(llm=llm_model,retriever=compression_retriever)
```

```python
result = chain.invoke("Why voldermord need harry potters help")
```

```python
from flashrank.Ranker import Ranker,RerankRequest
```

```python
def get_result (query,passsages,choice) :
  if choice == "Nano" :
    ranker = Ranker()
  elif choice == "Small" :
    ranker = Ranker(model_name='ms-marco-MiniLM-L-12-v2' , cache_dir = "/opt")
  elif choice == "Medium" :
    ranker = Ranker(model_name="rank-T5-flan",cache_dir = "/opt")
  elif choice == "Large" :
    ranker = Ranker(model_name="rank_zephyr_7b_v1_full" , cache_dir='/opt')
  rerankrequest = RerankRequest(query=query,passages=passages)
  result = ranker.rerank(rerankrequest)
  return result
```
